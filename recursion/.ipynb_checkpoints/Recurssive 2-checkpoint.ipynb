```python
def draw(leng):
    if leng > 10:
        print ('>',leng)
        dash.forward(leng)
        dash.left(30)
        print('>>',leng)
        draw(leng*.7)
        dash.right(60)
        print('>>>',leng)
        draw(leng*.7)
        dash.left(30)
        dash.backward(leng)
```
