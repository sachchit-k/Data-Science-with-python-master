## Contouring and pseudocolor

A pseudocolor plot is a rectangular array of cells with colors determined by C . MATLAB creates a pseudocolor plot by using each set of four adjacent points in C to define a surface patch (i.e., cell). pcolor(C) draws a pseudocolor plot. The elements of C are linearly mapped to an index into the current colormap.

```python
matInfo=np.random.random(100).reshape(10,10)
```

```python
matInfo.max(),matInfo.min()
```

```python
plt.figure(figsize=(8,8))
plt.pcolor(matInfo)
xx=matInfo.ravel()
for k in range(10):
    for l in range(10):
        plt.text(k,l,str(format(matInfo[k][l],'.1f')))
```

```python
plt.figure(figsize=(8,8))
cmap = plt.get_cmap('PiYG')
plt.pcolormesh(matInfo,cmap=cmap,shading='gouraud')
xx=matInfo.ravel()
# for k in range(10):
#     for l in range(10):
#         plt.text(k,l,str(format(matInfo[k][l],'.1f')))
```

## Pie Chart

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'A', 'B', 'C', 'D'
sizes = [15, 30, 45, 10]
explode = (0, 0.1,.1,.2)  # only "explode" the 2nd slice (i.e. 'Hogs')
plt.figure(figsize=(8,8))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')
```

## Polar graphs

```python
rangeOfValues = np.arange(0, 2, 0.1)
thetaVals = 2 * np.pi * rangeOfValues
```

```python
rangeOfValues
```

```python
plt.figure(figsize=(8,8))
plt.subplot(111, projection='polar')
plt.plot(thetaVals)
```

```python
ax = plt.subplot(111, projection='polar')
ax.plot(thetaVals, rangeOfValues)
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
```

## 3D plot

Demonstrates plotting a 3D surface colored with the coolwarm color map.

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
xval=np.arange(5)
yVal=np.arange(5,10)
```

```python
X,Y=np.meshgrid(xval,yVal)
```

```python
X
```

```python
Y
```

```python
X = np.arange(-3, 3, 0.25)
Y = np.arange(-3, 3, 0.30)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.gca(projection='3d') # Get current axes
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,linewidth=0, antialiased=False)
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.gca(projection='3d') # Get current axes
surf = ax.plot_surface(X, Y, Z, cmap=cm.winter,linewidth=0, antialiased=False)
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.gca(projection='3d') # Get current axes
surf = ax.plot_surface(X, Y, Z, cmap=cm.autumn,linewidth=0, antialiased=False)
plt.show()
```
